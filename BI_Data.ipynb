```python
def populate_phase_counts(original_df, result_df):
    # Iterate through the original DataFrame
    for _, row in original_df.iterrows():
        drug_id = row["DrugBank_ID"]
        phase = row["phase"]
        
        # Check if this drug_id is in our index
        if drug_id in result_df.index:
            # Determine which phase column to increment
            if phase == 1 or phase == "1" or phase == "Phase 1":
                result_df.loc[drug_id, "Phase 1"] += 1
            elif phase == 2 or phase == "2" or phase == "Phase 2":
                result_df.loc[drug_id, "Phase 2"] += 1
            elif phase == 3 or phase == "3" or phase == "Phase 3":
                result_df.loc[drug_id, "Phase 3"] += 1
    
    return result_df
```

```python

def phaseCount(DF, DBI):
    countP1 = 0
    countP2 = 0
    countP3 = 0
    for index, row in DF.iterrows():
        if DBI in str(row["DrugBank_ID"]) and "Phase 1" in row["phase"]:
            countP1 += 1
        if DBI in str(row["DrugBank_ID"]) and "Phase 2" in row["phase"]:
            countP2 += 1
        if DBI in str(row["DrugBank_ID"]) and "Phase 3" in row["phase"]:
            countP3 += 1
    return (countP1, countP2,countP3)
```
