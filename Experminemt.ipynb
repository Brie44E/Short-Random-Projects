```python

def gate1(investmentPer, initial, probGate1):
    #Gate1
    probGate1 = 0.5
    # Generate a random number between 0 and 1
    random_number = random.random()

    # Outcome based on probability of 0.5
    if random_number < probGate1:
        payoff = (investment/0.5)+cash
        print("Gate 1: Success; Payoff:", payoff)
        status1 = True
    else:
        payoff = (investment*0.5)+cash
        print("Gate 1: Failure; Payoff:", payoff)
        status1 = False

    return(payoff, status1)
```

```python
import itertools

class TreeNode:
    def __init__(self, name, probability, invest_percentage=None):
        self.name = name
        self.probability = probability
        self.invest_percentage = invest_percentage
        self.children = []

def calculate_outcomes(node, current_prob=1.0, current_money=100):
    if not node.children:
        return [(node.name, current_prob, current_money)]
    
    outcomes = []
    #invest_amount = current_money * node.invest_percentage
    invest_amount = current_money * (node.invest_percentage or 0.0)
    save_amount = current_money - invest_amount
    
    for child in node.children:
        if child.name.endswith("Success"):
            new_money = save_amount + (invest_amount * 2)
        else:
            new_money = save_amount + (invest_amount / 2)
        
        outcomes.extend(calculate_outcomes(child, 
                                           current_prob * child.probability, 
                                           new_money))
    return outcomes

def optimize_tree(node, step=0.1):
    if not node.children:
        return

    best_ev = 0
    best_percentage = 0

    for percentage in [i * step for i in range(11)]:  # 0 to 1 in steps of 0.1
        node.invest_percentage = percentage
        outcomes = calculate_outcomes(node)
        expected_value = sum(prob * money for _, prob, money in outcomes)
        
        if expected_value > best_ev:
            best_ev = expected_value
            best_percentage = percentage

    node.invest_percentage = best_percentage
    
    for child in node.children:
        optimize_tree(child, step)

def print_tree(node, level=0):
    if node.invest_percentage is None:
        percentage_str = "N/A"
    else:
        percentage_str = f"{node.invest_percentage:.1%}"
    print("  " * level + f"{node.name}: Invest {percentage_str}")
    for child in node.children:
        print_tree(child, level + 1)

def print_outcomes(outcomes):
    print("\nAll possible outcomes:")
    for path, probability, money in outcomes:
        print(f"Path: {path}, Probability: {probability:.4f}, Final Amount: ${money:.2f}")

def find_best_outcome(outcomes):
    best_outcome = max(outcomes, key=lambda x: x[2])
    print(f"\nBest outcome:")
    print(f"Path: {best_outcome[0]}, Probability: {best_outcome[1]:.4f}, Final Amount: ${best_outcome[2]:.2f}")

# Create the probability tree
root = TreeNode("Root", 1.0)

# Level 1
level1_a = TreeNode("1A", 0.5)
level1_b = TreeNode("1B", 0.5)
root.children = [level1_a, level1_b]

# Level 2
level2_aa = TreeNode("2AA Success", 0.667)
level2_ab = TreeNode("2AB Failure", 0.333)
level1_a.children = [level2_aa, level2_ab]

level2_ba = TreeNode("2BA Success", 0.5)
level2_bb = TreeNode("2BB Failure", 0.5)
level1_b.children = [level2_ba, level2_bb]

# Level 3
level3_aaa = TreeNode("3AAA Success", 0.333)
level3_aab = TreeNode("3AAB Failure", 0.667)
level2_aa.children = [level3_aaa, level3_aab]

level3_aba = TreeNode("3ABA Success", 0.333)
level3_abb = TreeNode("3ABB Failure", 0.667)
level2_ab.children = [level3_aba, level3_abb]

level3_baa = TreeNode("3BAA Success", 0.333)
level3_bab = TreeNode("3BAB Failure", 0.667)
level2_ba.children = [level3_baa, level3_bab]

level3_bba = TreeNode("3BBA Success", 0.5)
level3_bbb = TreeNode("3BBB Failure", 0.5)
level2_bb.children = [level3_bba, level3_bbb]

# Optimize the tree
optimize_tree(root)

# Print the optimized tree
print("Optimized Investment Percentages:")
print_tree(root)

# Calculate and print outcomes
outcomes = calculate_outcomes(root)
print_outcomes(outcomes)
find_best_outcome(outcomes)
```
