```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
with open('/usr/share/dict/words', 'r') as file:
    words = file.read().splitlines()
```

```python
# Count the occurrences of each letter
all_letters = ''.join(words).lower()  # Combine all words into a single string and convert to lowercase
letter_counts = Counter(all_letters)  # Count each letter

# Create a Pandas DataFrame
df = pd.DataFrame(list(letter_counts.items()), columns=['Letter', 'Frequency']).sort_values(by='Frequency', ascending=False)
```

```python
# Display the DataFrame
print(df)
```

```python
# Plot the frequencies as a bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Letter'], df['Frequency'], color='skyblue')
plt.title('Frequency of Letters', fontsize=16)
plt.xlabel('Letter', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
