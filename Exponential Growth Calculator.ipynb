```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def returnCalc(interest, base = 1000, goal = 1000000):

    # Set up data
    x = np.linspace(0, 10, 100)
    y = np.sin(x)

    xValues = []
    yValues = []

    orBase = base

    periods = 0
    while base <= goal: #1 million
        xValues.append(periods)
        yValues.append(base)
        base = base * (1+interest)
        periods += 1

    plt.title("Exponantial Growth")
    plt.plot(xValues, yValues)
    plt.show()

    print(f"With a interest of {interest} it takes {periods} periods to come from {orBase} to {base}")
```

```python
interest = 0.1
returnCalc(interest)
```
